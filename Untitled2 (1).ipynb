```python
import re
import shutil
from pathlib import Path

DATA_ROOT = Path("/content/DATA_CLEAN")
IMAGES_DIR = DATA_ROOT / "images"
LABELS_DIR = DATA_ROOT / "labels"

# Handle possible double-nesting like images/images/
if (IMAGES_DIR / "images").exists():
    IMAGES_DIR = IMAGES_DIR / "images"
if (LABELS_DIR / "labels").exists():
    LABELS_DIR = LABELS_DIR / "labels"

print(f"Images from: {IMAGES_DIR}")
print(f"Labels from: {LABELS_DIR}")

# Regex to detect (N) suffix in filenames
img_dup = re.compile(r"\(\d+\)$")
lbl_dup = re.compile(r"\(\d+\)\.txt$")

# Find original (non-duplicate) files
image_files = list(IMAGES_DIR.glob("*.jpg"))
label_files = [f for f in LABELS_DIR.glob("*.txt") if not lbl_dup.search(f.name)]

# Use only unsuffixed image names; match against label stems
label_stems = {f.stem for f in label_files}
matched = []
seen = set()
for img in image_files:
    if img_dup.search(img.stem):
        continue  # skip (N) duplicates
    if img.stem in seen:
        continue
    if img.stem in label_stems:
        matched.append((img, LABELS_DIR / f"{img.stem}.txt"))
        seen.add(img.stem)

print(f"Matched image-label pairs: {len(matched)}")

# Build clean folder structure
CLEAN = Path("/content/clean")
if CLEAN.exists():
    shutil.rmtree(CLEAN)
(CLEAN / "images").mkdir(parents=True)
(CLEAN / "labels").mkdir(parents=True)

for img_path, label_path in matched:
    shutil.copy2(img_path, CLEAN / "images" / img_path.name)
    shutil.copy2(label_path, CLEAN / "labels" / label_path.name)

n_imgs = len(list((CLEAN / "images").iterdir()))
n_lbls = len(list((CLEAN / "labels").iterdir()))
print(f"Clean dataset built: {n_imgs} images, {n_lbls} labels")
```

```python
import os
import random
import glob
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.patches as patches
from pathlib import Path
from PIL import Image
from ultralytics import YOLO

# Set random seed for reproducibility
random.seed(42)
np.random.seed(42)

# Point directly to your Colab directory tree layout
DATA_DIR = Path("/content/DATA_CLEAN")

print(f"Data directory target verified: {DATA_DIR.exists()}")
```

```python
# Aggressive recursive search to find images inside images/ or anywhere under DATA_CLEAN
image_extensions = ('*.jpg', '*.jpeg', '*.png', '*.JPG', '*.JPEG', '*.PNG')
image_files = []
for ext in image_extensions:
    image_files.extend(glob.glob(str(DATA_DIR / "**" / ext), recursive=True))

label_files = glob.glob(str(DATA_DIR / "**" / "*.txt"), recursive=True)

# Clean up split lists from text tracking if they accidentally get matched
label_files = [f for f in label_files if not any(k in os.path.basename(f) for k in ['train.txt', 'val.txt', 'test.txt'])]

print(f"Total Images Found: {len(image_files)}")
print(f"Total Label Files Found: {len(label_files)}")

# Check alignment via file base names
image_basenames = {os.path.splitext(os.path.basename(f))[0] for f in image_files}
label_basenames = {os.path.splitext(os.path.basename(f))[0] for f in label_files}

missing_labels = image_basenames - label_basenames
missing_images = label_basenames - image_basenames

print(f"Images missing labels: {len(missing_labels)}")
print(f"Labels missing images: {len(missing_images)}")

# Class distribution analysis
class_counts = {}
for lbl_path in label_files:
    if os.path.exists(lbl_path):
        with open(lbl_path, 'r') as f:
            for line in f:
                parts = line.strip().split()
                if parts:
                    cls_id = int(parts[0])
                    class_counts[cls_id] = class_counts.get(cls_id, 0) + 1

print("\n--- Per-Class Object Distribution ---")
df_classes = pd.DataFrame(list(class_counts.items()), columns=['Class ID', 'Object Count']).sort_values('Class ID')
print(df_classes.to_string(index=False))

# Image size statistics (sampling up to 100 images securely)
sampled_images = random.sample(image_files, min(100, len(image_files)))
widths, heights = [], []
for img_path in sampled_images:
    with Image.open(img_path) as img:
        w, h = img.size
        widths.append(w)
        heights.append(h)

print("\n--- Image Dimension Statistics ---")
print(f"Width  -> Min: {min(widths)}, Max: {max(widths)}, Mean: {np.mean(widths):.1f}")
print(f"Height -> Min: {min(heights)}, Max: {max(heights)}, Mean: {np.mean(heights):.1f}")
```

```python
# Clean up absolute file paths for splitting
all_images_clean = [os.path.abspath(p) for p in image_files]
random.seed(42)
random.shuffle(all_images_clean)

total = len(all_images_clean)
train_idx = int(total * 0.70)
val_idx = train_idx + int(total * 0.15)

train_splits = all_images_clean[:train_idx]
val_splits = all_images_clean[train_idx:val_idx]
test_splits = all_images_clean[val_idx:]

# Save split descriptors directly into the path folder where YOLO looks
with open(str(DATA_DIR / "train.txt"), "w") as f:
    f.write("\n".join(train_splits))
with open(str(DATA_DIR / "val.txt"), "w") as f:
    f.write("\n".join(val_splits))
with open(str(DATA_DIR / "test.txt"), "w") as f:
    f.write("\n".join(test_splits))

print(f"Splits complete: Train={len(train_splits)}, Val={len(val_splits)}, Test={len(test_splits)}")
```

```python
# Load pre-trained weights for selected variant
model = YOLO("yolo26s.pt")

# Initiate fully convolutional training routine
results = model.train(
    data="/content/DATA_CLEAN/data.yaml",  # <-- Give it the full path here!
    epochs=24,
    imgsz=640,
    batch=8,
    project="runs",
    name="cats_v1",
    seed=42,
)
```

```python
# Display training results summary chart
results_png_path = "runs/cats_v1/results.png"
if os.path.exists(results_png_path):
    img = Image.open(results_png_path)
    plt.figure(figsize=(12, 12))
    plt.imshow(img)
    plt.axis('off')
    plt.show()

# Read metrics from csv to confirm absolute maximums
results_csv = "runs/cats_v1/results.csv"
if os.path.exists(results_csv):
    df = pd.read_csv(results_csv)
    df.columns = df.columns.str.strip()
    best_idx = df['metrics/mAP50(B)'].idxmax()
    print(f"Best mAP@0.5: {df.loc[best_idx, 'metrics/mAP50(B)'].round(4)} reached at Epoch {best_idx + 1}")
    print(f"Best mAP@0.5:0.95: {df.loc[best_idx, 'metrics/mAP50-95(B)'].round(4)}")
```

```python
# Force validation pass targeting the isolated test split configuration
metrics = model.val(data="/content/DATA_CLEAN/data.yaml", split="test")

# Package performance output vectors into a table
metrics_data = {
    "Metric": ["mAP@0.5", "mAP@0.5:0.95", "Mean Precision (P)", "Mean Recall (R)"],
    "Value": [metrics.box.map50, metrics.box.map, metrics.box.mp, metrics.box.mr]
}
df_metrics = pd.DataFrame(metrics_data)
print(df_metrics.to_string(index=False))
```

```python
# Collect test split images from saved text registry
with open("/content/DATA_CLEAN/test.txt", "r") as f:
    test_img_paths = [line.strip() for line in f if line.strip()]

# Sample 6 distinct images for qualitative assessment
test_samples = random.sample(test_img_paths, 6)
results_list = model.predict(source=test_samples, conf=0.25, verbose=False)

fig, axes = plt.subplots(2, 3, figsize=(18, 12))

for idx, res in enumerate(results_list):
    ax = axes[idx // 3, idx % 3]
    orig_img = cv2.cvtColor(res.orig_img, cv2.COLOR_BGR2RGB)
    ax.imshow(orig_img)
    h, w, _ = orig_img.shape

    # Trace Predicted Bounding Configurations
    for box in res.boxes:
        xyxy = box.xyxy[0].cpu().numpy()
        conf = box.conf[0].cpu().item()
        cls_val = int(box.cls[0].cpu().item())

        pred_rect = patches.Rectangle((xyxy[0], xyxy[1]), xyxy[2]-xyxy[0], xyxy[3]-xyxy[1],
                                      linewidth=2, edgecolor='cyan', facecolor='none')
        ax.add_patch(pred_rect)
        ax.text(xyxy[0], xyxy[1]-2, f"P: {res.names[cls_val]} {conf:.2f}",
                color='white', fontsize=9, weight='bold', backgroundcolor='cyan')

    # Overlap True Anchors for the first 3 tracking windows
    if idx < 3:
        base_name = os.path.splitext(os.path.basename(test_samples[idx]))[0]
        lbl_p = DATA_DIR / "labels" / f"{base_name}.txt" if (DATA_DIR / "labels").exists() else DATA_DIR / f"{base_name}.txt"
        if os.path.exists(lbl_p):
            with open(lbl_p, 'r') as f:
                for line in f:
                    c_id, cx, cy, box_w, box_h = map(float, line.strip().split())
                    abs_w = box_w * w
                    abs_h = box_h * h
                    abs_x = (cx * w) - (abs_w / 2)
                    abs_y = (cy * h) - (abs_h / 2)
                    gt_rect = patches.Rectangle((abs_x, abs_y), abs_w, abs_h,
                                                linewidth=2, edgecolor='magenta', facecolor='none', linestyle='--')
                    ax.add_patch(gt_rect)
                    ax.text(abs_x, abs_y - 15, "GT", color='white', fontsize=8, weight='bold', backgroundcolor='magenta')

    ax.set_title(f"Sample Image {idx+1}")
    ax.axis('off')

plt.tight_layout()
plt.show()
```
